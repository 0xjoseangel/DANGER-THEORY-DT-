# Danger Theory 

## El Mecanismo Europeo de Cambio (MEC) y la "Gran Divergencia"

Antes de la creación del Euro, la mayoría de los países de la Comunidad Económica Europea (CEE) formaban parte del **Mecanismo Europeo de Cambio (MEC)**, fundado en 1979.

### ¿Qué era el MEC y para qué servía?

El objetivo principal del MEC era **reducir la volatilidad** entre las monedas europeas y **fomentar la estabilidad** económica. Se consideraba un paso fundamental para crear un mercado único y, eventualmente, una moneda única (el Euro).

Funcionaba así:

1.  **Ancla del Sistema:** El **Marco Alemán (DEM)** era el "ancla" no oficial del sistema. Debido a la fortaleza y baja inflación de la economía alemana, las demás monedas buscaban mantener un valor estable frente al DEM.
2.  **Bandas de Flctuación:** Cada moneda tenía un tipo de cambio central fijado frente a las demás. Solo se les permitía fluctuar dentro de una "banda" estrecha (generalmente ±2.25% o ±6%) alrededor de ese tipo central.
3.  **Intervención Obligatoria:** Si una moneda se debilitaba y se acercaba al límite inferior de su banda, su banco central estaba **obligado a intervenir**: debía **comprar** su propia moneda (gastando sus reservas de divisas, como el DEM) para subirle el precio. A la inversa, los bancos centrales de los países con monedas fuertes debían vender su moneda para bajarle el precio.

### El Problema: El Miércoles Negro (Septiembre 1992)

Este sistema se volvió insostenible a principios de los 90. Tras la reunificación alemana, Alemania subió sus tipos de interés para controlar la inflación. Esto hizo al DEM *aún más fuerte* y atrajo capital, debilitando a las demás monedas.

Países como el Reino Unido (GBP) e Italia (ITL) sufrían una recesión y **no podían permitirse subir sus tipos de interés** para igualar a Alemania (eso habría hundido más sus economías).

Los especuladores (como George Soros) vieron esta debilidad y apostaron miles de millones a que el Reino Unido e Italia no podrían seguir gastando sus reservas para defender un tipo de cambio artificialmente alto.

Tenían razón. El **16 de septiembre de 1992 (el "Miércoles Negro")**, el Reino Unido fue forzado a abandonar el MEC, y la libra se desplomó. Italia le siguió poco después.

Podemos dividir el gráfico en tres fases claras:

1. Antes de la caída (Ene 1991 - Sep 1992)

Valor alto y estable: La libra se mantiene fuerte y relativamente estable. Su valor fluctúa en un rango estrecho, mayormente entre 2.9 y 3.0 marcos alemanes.

Contexto: Durante este tiempo, el Reino Unido formaba parte del Mecanismo Europeo de Cambio (MEC), que intentaba mantener los tipos de cambio fijos para estabilizar las monedas antes de la creación del Euro. El marco alemán era la moneda "ancla" del sistema.

2. El colapso (Septiembre de 1992)

Caída vertical: En un período muy corto, el valor de la libra se desploma. Cae desde aproximadamente 2.8 marcos hasta un mínimo de alrededor de 2.3 marcos.

Significado: Esto representa una devaluación masiva de la libra (perdió cerca del 20% de su valor frente al marco).

Contexto (El Miércoles Negro): Los especuladores (el más famoso fue George Soros) apostaron miles de millones a que el gobierno británico no podría mantener artificialmente el valor de la libra tan alto como exigía el MEC. Tenían razón. El gobierno británico intentó defender la libra, pero fracasó y tuvo que retirarse del MEC, dejando que la moneda cayera a su valor "real" de mercado.

3. Después de la caída (Finales de 1992 - Ene 1995)

Una "nueva normalidad" más baja: La libra nunca recupera su valor anterior. Se estabiliza en un nuevo rango, mucho más bajo.

Volatilidad: El tipo de cambio sigue siendo volátil (sube y baja mucho), pero ahora se mueve en un canal entre 2.3 y 2.6 marcos, muy por debajo de los 2.9-3.0 anteriores.

El gráfico es la prueba de la intervención coordinada y fallida de los bancos centrales europeos para salvar a la libra.

El contexto clave sigue siendo el Mecanismo Europeo de Cambio (MEC). Este mecanismo no solo obligaba al Reino Unido a defender la libra; también obligaba a otros miembros (como el Bundesbank de Alemania) a intervenir si una moneda alcanzaba su banda de fluctuación.

Si la Libra (GBP) se debilitaba (lo que estaba pasando), el BoE tenía que comprar libras usando sus reservas de Marcos (DEM).

Al mismo tiempo, el Bundesbank (Alemania) estaba obligado a comprar libras usando sus propios marcos.

De nuevo, dividamos el gráfico en partes para su análisis

1. El Pico: La Defensa (Mediados de 1992)
Lo que vemos: Un aumento masivo y vertical de las reservas de GBP en manos extranjeras, alcanzando un pico de más de 48 mil millones justo antes del 16 de septiembre.

Análisis Económico: Esto no es un signo de confianza. Es todo lo contrario.

Significado: Es la evidencia de la intervención masiva. A medida que los especuladores vendían miles de millones de libras, los bancos centrales del MEC (principalmente el Bundesbank) estaban en el otro lado de la operación, comprándolas forzosamente para intentar que el precio no cayera.

Cada libra que compraba el Bundesbank se añadía a sus "Reservas Internacionales de GBP". El pico en el gráfico muestra la cantidad desesperada de libras que otros países se vieron forzados a absorber para cumplir con el pacto del MEC.

2. El Desplome: El Pánico (Después del 16 de Septiembre de 1992)
Lo que vemos: En el momento en que el Reino Unido abandona el MEC (la línea verde), las reservas de GBP en manos extranjeras se desploman, cayendo desde 48 mil millones a casi 41 mil millones en muy poco tiempo.

Análisis Económico: El 16 de septiembre, el Reino Unido se rinde. La libra entra en caída libre (como vimos en tu primer gráfico).

Significado: Todos esos bancos centrales (Alemania, Francia, etc.) que acababan de acumular 48 mil millones en libras, ahora están sentados sobre un activo que se está devaluando por minutos. El pacto del MEC está roto, por lo que ya no tienen obligación de retenerlas.

Inicia un "sálvese quien pueda". Estos bancos centrales empiezan a vender masivamente (dumpear) sus nuevas y tóxicas reservas de libras para limitar sus propias pérdidas financieras.



### El Gráfico: Visualizando la Ruptura

A continuacion analizaremos varias monedas para ver como funcionaba el MEC

Para ver esta historia, no podemos simplemente graficar los precios (una lira vale 1200 y una libra 1.6; las escalas son muy distintas).

La solución es **normalizar** los datos: elegimos una fecha de inicio (1 de enero de 1990) y fijamos el valor de todas las monedas en "100" en ese día. El gráfico mostrará el **crecimiento (o caída) porcentual** de cada moneda desde ese punto.

**Qué veremos:**
* **Antes de 1992:** Las líneas del DEM, ITL y ESP deberían moverse muy juntas, mostrando que el MEC las mantenía "atadas".
* **En Septiembre de 1992:** Veremos una violenta **divergencia**: las líneas de GBP e ITL se desplomarán, mientras que el DEM se mantendrá fuerte. Esta es la ruptura del sistema.

```python
# --- 1. Carga y Limpieza de Datos ---

# Ajusta el 'skiprows=3' para que coincida con tu archivo Excel.
# Esto asume que la fila con 'Dates', 'DEM Curncy' es la fila 4.
excel_file_path = 'data/tipoDeCambioEuropa.xlsx'
df = pd.read_excel(excel_file_path, skiprows=3)

print(df)
```

Como el excel tiene un formato complejo hay que limpiar los datos

```python
# 1.1. Seleccionar solo las filas de datos (desde la fila 3 en adelante)
# Usamos .copy() para evitar un 'SettingWithCopyWarning'
df_cleaned = df.iloc[3:].copy()

# 1.2. Asignar los nombres de columna correctos
# Basado en tu captura:
# Unnamed: 0 = Basura
# Unnamed: 1 = Dates
# Unnamed: 2 = DEM Curncy
# Unnamed: 3 = ITL Curncy
# Unnamed: 4 = ESP Curncy
# Unnamed: 5 = GBP Curncy
df_cleaned.columns = ['Basura', 'Date', 'DEM', 'ITL', 'ESP', 'GBP']

# 1.3. Eliminar la primera columna 'Basura' que está llena de 'NaN'
df_cleaned = df_cleaned.drop(columns=['Basura'])

# 1.4. Convertir la columna de fecha a formato datetime
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])

# 1.5. Limpiar las columnas de moneda (convertir a número)
# (Tu 'print(df)' muestra puntos decimales, pero el Excel original
# mostraba comas, así que este paso es más seguro)
for col in ['DEM', 'ITL', 'ESP', 'GBP']:
    # Convierte a string, reemplaza comas por si acaso, y convierte a numérico
    df_cleaned[col] = df_cleaned[col].astype(str).str.replace(',', '.')
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# 1.6. Eliminar filas donde falten datos (si los hay)
df_cleaned = df_cleaned.dropna()

# 1.7. Establecer la fecha como el índice (index)
df_cleaned = df_cleaned.set_index('Date').sort_index()

print("--- Datos Limpios (primeras 5 filas) ---")
print(df_cleaned.head())
print("\n--- Información del DataFrame ---")
df_cleaned.info()
```

Ahora si, graficamos

```python
# --- Normalización de los Datos ---

# Dividimos cada valor por el valor de la primera fila (iloc[0]) y multiplicamos por 100
# Esto hace que todas las monedas empiecen en un valor base de 100
df_normalized = df_cleaned.div(df_cleaned.iloc[0]) * 100

print("\n--- Datos Normalizados (primeras 5 filas) ---")
print(df_normalized.head())


plt.figure(figsize=(14, 8))

# ... (aquí van todas tus líneas `plt.plot()`) ...
plt.plot(df_normalized['DEM'], label='Marco Alemán (DEM)')
plt.plot(df_normalized['ITL'], label='Lira Italiana (ITL)')
plt.plot(df_normalized['ESP'], label='Peseta Española (ESP)')
plt.plot(df_normalized['GBP'], label='Libra Esterlina (GBP)')


# ... (aquí van tus líneas `plt.axhline()` y `plt.axvline()`) ...
plt.axhline(100, color='black', linestyle='--', linewidth=1, label='Valor Inicial (100)')
plt.axvline(pd.to_datetime('1992-09-16'), color='red', linestyle=':', linewidth=2, label='Miércoles Negro (Sep 16, 1992)')

# --- CONFIGURACIÓN Y ESTILO ---
plt.title('La Gran Divergencia: Ruptura del Mecanismo Europeo de Cambio (MEC)', fontsize=16)
plt.ylabel('Valor Normalizado (Base 100 = 01-Ene-1990)')
plt.xlabel('Fecha')

# --- ✅ AQUÍ ACOTAMOS EL EJE X ---
plt.xlim(pd.to_datetime('1990-01-01'), pd.to_datetime('1995-01-01'))

# ... (aquí va el resto: legend, grid, tight_layout, etc.) ...
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# ... (aquí va la configuración de los ejes) ...
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=[1, 7]))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.show()
```

Una observación clave del gráfico es la alta volatilidad en 1991, que parece contradecir la idea de "estabilidad" del MEC. Lo que vemos no es calma, sino **los síntomas y tensiones** que llevaron directamente al colapso de 1992.

Para entender este período, primero hay que recordar cómo se leen los datos (que hemos normalizado):

* **Para la Libra (GBP, roja):** La cotización es `USD/GBP`. Por lo tanto, un valor **más alto** significa una **Libra más fuerte**.
* **Para las otras (DEM, ITL, ESP):** La cotización es `Moneda/USD`. Un valor **más alto** significa una **moneda más débil** (se necesitan más liras/pesetas por dólar).

Con esto en mente, la volatilidad de 1991 se explica en tres fases:

### 1. La "Burbuja" de la Libra (Principios de 1991)

* **Qué vemos:** La línea roja (GBP) se dispara **hacia arriba** (se fortalece enormemente), mientras que las otras monedas (azul, naranja, verde) se mueven en la dirección opuesta.
* **Análisis Económico:** El Reino Unido acababa de entrar en el MEC (Octubre 1990) en un intento de frenar su alta inflación interna. Para hacerlo, el gobierno fijó unos **tipos de interés altísimos (cerca del 15%)**.
* **Consecuencia:** Los inversores globales vieron una oportunidad de oro: comprar libras esterlinas (que ahora estaban "seguras" y "estables" gracias al MEC) y, al mismo tiempo, recibir un interés altísimo. Esta entrada masiva de capital **fortaleció artificialmente la libra**, creando el pico que vemos.


### 2. La Corrección: Recesión y Tensión (Mediados 1991 - 1992)

* **Qué vemos:** La línea roja (GBP) empieza a **caer** (debilitarse) desde su pico artificial. Las otras monedas (ITL, ESP) empiezan a **subir** (debilitarse), y todas parecen "juntarse" de nuevo.
* **Análisis Económico:** Este no es un retorno a la calma, es el inicio del problema.
    * **Recesión en UK:** Esos tipos de interés del 15% estaban **asfixiando la economía británica**, que entró en una recesión profunda. Era insostenible.
    * **Fuerza Alemana:** Al mismo tiempo, Alemania (línea azul, DEM), mantenía sus propios tipos de interés altos para financiar su reunificación, lo que fortalecía al marco.
    * **El "Ajuste":** Lo que vemos como "juntarse" es en realidad la **corrección de la burbuja** de la libra (que caía a su valor real) y la **creciente presión** sobre la lira y la peseta, que no podían seguir el ritmo del poderoso marco alemán.

### 3. La Separación: El Colapso (Septiembre 1992)

* **Qué vemos:** La línea roja punteada. El sistema se rompe.
* **Análisis Económico:** Los especuladores vieron esta contradicción: la economía del Reino Unido *necesitaba* tipos de interés bajos y una libra débil (más baja en el gráfico) para salir de la recesión, pero el MEC le *obligaba* a mantener tipos altos y una libra fuerte.
* **Consecuencia:** Apostaron a que el Reino Unido no aguantaría. Acertaron. La libra (roja) se desploma (se devalúa) y la lira (naranja) y peseta (verde) se disparan (se devalúan), marcando el fin del sistema tal y como se conocía.
