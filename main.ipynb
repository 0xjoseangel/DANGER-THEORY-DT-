```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de grficos
sns.set_theme(style="whitegrid")

# 1. Crear datos simulados de una crisis
# Fechas: 100 das antes y despus de una "crisis" terica
fechas = pd.date_range(start="1992-06-01", periods=200, freq='B') # 'B' son das hbiles (Business days)
# Simulamos un tipo de cambio estable que de repente cae
tipo_cambio = np.concatenate([
    np.random.normal(2.95, 0.02, 150),  # 150 das estables alrededor de 2.95 DM/
    np.random.normal(2.40, 0.05, 50)    # 50 das tras la devaluacin a ~2.40
])

# Crear DataFrame
df_simulado = pd.DataFrame({'Fecha': fechas, 'Tipo_Cambio_DM_GBP': tipo_cambio})
df_simulado.set_index('Fecha', inplace=True)

# 2. Graficar
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_simulado, x=df_simulado.index, y='Tipo_Cambio_DM_GBP', color='navy', linewidth=2)

# Aadir detalles al grfico
plt.title('Simulacin de Prueba: Colapso del Tipo de Cambio', fontsize=14)
plt.ylabel('Marcos Alemanes por Libra ()', fontsize=12)
plt.xlabel('Fecha', fontsize=12)
plt.axvline(x=df_simulado.index[150], color='red', linestyle='--', label='Inicio Crisis Simulada')
plt.legend()

plt.show()

print(" Si ves el grfico arriba, tu entorno 'dt_project' est listo para trabajar.")
```
