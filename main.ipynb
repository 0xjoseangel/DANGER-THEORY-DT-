# Danger Theory 

## El Mecanismo Europeo de Cambio (MEC) y la "Gran Divergencia"

Antes de la creación del Euro, la mayoría de los países de la Comunidad Económica Europea (CEE) formaban parte del **Mecanismo Europeo de Cambio (MEC)**, fundado en 1979.

### ¿Qué era el MEC y para qué servía?

El objetivo principal del MEC era **reducir la volatilidad** entre las monedas europeas y **fomentar la estabilidad** económica. Se consideraba un paso fundamental para crear un mercado único y, eventualmente, una moneda única (el Euro).

Funcionaba así:

1.  **Ancla del Sistema:** El **Marco Alemán (DEM)** era el "ancla" no oficial del sistema. Debido a la fortaleza y baja inflación de la economía alemana, las demás monedas buscaban mantener un valor estable frente al DEM.
2.  **Bandas de Flctuación:** Cada moneda tenía un tipo de cambio central fijado frente a las demás. Solo se les permitía fluctuar dentro de una "banda" estrecha (generalmente ±2.25% o ±6%) alrededor de ese tipo central.
3.  **Intervención Obligatoria:** Si una moneda se debilitaba y se acercaba al límite inferior de su banda, su banco central estaba **obligado a intervenir**: debía **comprar** su propia moneda (gastando sus reservas de divisas, como el DEM) para subirle el precio. A la inversa, los bancos centrales de los países con monedas fuertes debían vender su moneda para bajarle el precio.

### El Problema: El Miércoles Negro (Septiembre 1992)

Este sistema se volvió insostenible a principios de los 90. Tras la reunificación alemana, Alemania subió sus tipos de interés para controlar la inflación. Esto hizo al DEM *aún más fuerte* y atrajo capital, debilitando a las demás monedas.

Países como el Reino Unido (GBP) e Italia (ITL) sufrían una recesión y **no podían permitirse subir sus tipos de interés** para igualar a Alemania (eso habría hundido más sus economías).

Los especuladores (como George Soros) vieron esta debilidad y apostaron miles de millones a que el Reino Unido e Italia no podrían seguir gastando sus reservas para defender un tipo de cambio artificialmente alto.

Tenían razón. El **16 de septiembre de 1992 (el "Miércoles Negro")**, el Reino Unido fue forzado a abandonar el MEC, y la libra se desplomó. Italia le siguió poco después.

* **Consecuencia:** Los inversores globales vieron una oportunidad de oro: comprar libras esterlinas (que ahora estaban "seguras" y "estables" gracias al MEC) y, al mismo tiempo, recibir un interés altísimo. Esta entrada masiva de capital **fortaleció artificialmente la libra**, creando el pico que vemos.

* **Análisis Económico:** Este no es un retorno a la calma, es el inicio del problema.

    * **El "Ajuste":** Lo que vemos como "juntarse" es en realidad la **corrección de la burbuja** de la libra (que caía a su valor real) y la **creciente presión** sobre la lira y la peseta, que no podían seguir el ritmo del poderoso marco alemán.

### 3. La Separación: El Colapso (Septiembre 1992)

* **Qué vemos:** La línea roja punteada. El sistema se rompe.
* **Análisis Económico:** Los especuladores vieron esta contradicción: la economía del Reino Unido *necesitaba* tipos de interés bajos y una libra débil (más baja en el gráfico) para salir de la recesión, pero el MEC le *obligaba* a mantener tipos altos y una libra fuerte.
* **Consecuencia:** Apostaron a que el Reino Unido no aguantaría. Acertaron. La libra (roja) se desploma (se devalúa) y la lira (naranja) y peseta (verde) se disparan (se devalúan), marcando el fin del sistema tal y como se conocía.

# 3. El Motor del Modelo: Algoritmo Evolutivo y Selección Natural

Hasta ahora hemos analizado el pasado (los datos históricos). Ahora, para validar la **Danger Theory**, no queremos simplemente mirar los gráficos nosotros mismos; queremos ver si una "inteligencia" artificial, que no sabe nada del futuro, podría haber predicho la crisis basándose únicamente en las señales disponibles.

Para lograrlo, utilizamos un **Algoritmo Evolutivo (Genetic Algorithm)**.

## 3.1 ¿Qué es un Algoritmo Evolutivo?

Imagina que no sabemos cuál es la fórmula perfecta para predecir una crisis. En lugar de intentar adivinarla matemáticamente, copiamos a la naturaleza. Utilizamos los principios de la **Teoría de la Evolución de Darwin** aplicados a la economía.

En nuestro simulador, creamos una población de **"Agentes Digitales"** (inversores artificiales).
* Al principio, son **tontos**: toman decisiones al azar.
* Los soltamos en el mercado de 1990-1992.
* Aquellos que pierden dinero o no ven la crisis, **"mueren"** (son eliminados).
* Aquellos que logran salvar su capital, **"sobreviven" y tienen "hijos"**.
* Sus hijos heredan sus estrategias, pero con pequeñas variaciones (**mutaciones**).

Tras repetir este proceso durante muchas generaciones, obtenemos un "Super Agente" que ha aprendido la estrategia óptima de supervivencia.


## 3.2 Anatomía de nuestros "Agentes"

Para que el ordenador entienda esto, traducimos la biología a código Python usando la librería `DEAP`. Así es como funciona la analogía:

### A. El ADN (Genotipo)
En los seres humanos, el ADN determina el color de ojos o la altura. En nuestros agentes, el ADN es una lista de **tres números (pesos)** que determinan a qué le prestan atención:

1.  **Gen de Reservas ($w_{res}$):** ¿Cuánto le importa al agente que el Banco de Inglaterra pierda reservas?
2.  **Gen de Precio ($w_{precio}$):** ¿Cuánto le importa el valor actual de la Libra?
3.  **Gen de Volatilidad ($w_{vol}$):** ¿Cuánto le asusta que el mercado se mueva mucho?

Un agente podría tener un ADN como `[0.1, 0.9, 0.0]`. Este agente sería un inversor tradicional: ignora las reservas y solo mira el precio.
Otro agente podría tener `[-0.8, 0.1, 0.5]`. Este agente está obsesionado con las reservas (Danger Theory).

### B. La Vida (El Entorno)
La "vida" del agente consiste en recorrer día a día nuestra tabla de datos `df_env` desde 1991 hasta 1992. Cada día, el agente calcula su **Señal de Alarma** usando su ADN:

$$Señal = (w_{res} \times Reservas) + (w_{precio} \times Precio) + (w_{vol} \times Volatilidad)$$

Si la señal supera un umbral de pánico (0.5), el agente **vende** sus libras y sale del mercado.

### C. La Supervivencia (Función de Fitness)
Esta es la parte más crítica. ¿Cómo decide el algoritmo quién es "apto" para vivir? Hemos diseñado una función de puntuación (**Fitness**) que premia la precisión:

* **Si vende muy pronto (ej. 1991):** Recibe pocos puntos. Ha sobrevivido, pero ha perdido mucho dinero en intereses por salirse del mercado antes de tiempo (falsa alarma).
* **Si vende muy tarde (después del 16 sep 1992):** **Castigo severo**. El agente se "comió" la crisis y perdió su capital.
* **Si vende justo antes (Agosto/Septiembre 1992):** **Puntuación Máxima**. Este agente detectó el peligro real y maximizó sus ganancias.

## 3.3 El Ciclo Evolutivo (La librería DEAP)

La librería `DEAP` se encarga de gestionar el "Círculo de la Vida" en nuestro código:

1.  **Población Inicial:** Creamos 100 agentes con ADNs totalmente aleatorios. La mayoría serán desastrosos.
2.  **Evaluación:** Soltamos a los 100 agentes en los datos históricos y les damos una nota (Fitness) según lo bien que lo hicieron.
3.  **Selección:** Elegimos a los mejores (los que tienen mejor nota) para ser padres. Usamos "Torneos": cogemos 3 al azar y el mejor de los 3 pasa a la siguiente ronda.
4.  **Cruce (Crossover):** Tomamos dos padres ganadores y mezclamos sus genes.
    * *Padre A (mira reservas)* + *Padre B (mira volatilidad)* = *Hijo C (mira reservas y volatilidad)*.
5.  **Mutación:** Con una pequeña probabilidad, cambiamos un gen al azar. Esto introduce "creatividad" o "innovación" en el sistema, permitiendo descubrir estrategias que a los padres no se les ocurrieron.

Este ciclo se repite 40 veces (Generaciones). Al final, el agente que queda es el resultado de una evolución dirigida a encontrar la verdad oculta en los datos.

Si nuestra **Danger Theory** es correcta, el agente ganador debería tener un **Gen de Reservas muy fuerte** y haber predicho la fecha exacta del Miércoles Negro.

```python
def prepare_real_data(df_rates, df_res, start_date='1990-01-01', end_date='1995-01-01'):
    rates = df_rates.copy().reset_index()
    reserves = df_res.copy()
    
    rates['Date'] = pd.to_datetime(rates['Date'])
    
    # ¡IMPORTANTE! Para las reservas usamos dayfirst=True por el formato (31/12/1994)
    reserves['Date'] = pd.to_datetime(reserves['Date'], dayfirst=True)
    
    # Ordenamos las reservas de más antigua a más reciente
    reserves = reserves.sort_values('Date')

    # Unimos usando 'left': Mantenemos TODAS las fechas del tipo de cambio (diario)
    # y pegamos las reservas donde coincida la fecha.
    df_env = pd.merge(rates, reserves, on='Date', how='left')
    
    # Ordenamos por fecha para que el relleno funcione
    df_env = df_env.sort_values('Date')

    # Si no hay dato de reservas se usa el ultimo conocido
    df_env['Total Reserves'] = df_env['Total Reserves'].ffill()

    mask = (df_env['Date'] >= start_date) & (df_env['Date'] <= end_date)
    df_env = df_env.loc[mask].reset_index(drop=True)

    # Si quedan NaNs al principio (antes del primer reporte de reservas), los borramos
    df_env = df_env.dropna(subset=['Total Reserves', 'GBP'])

    # GENERAR GENES (Normalización)
    df_env['Reserves_Norm'] = (df_env['Total Reserves'] - df_env['Total Reserves'].mean()) / df_env['Total Reserves'].std()
    df_env['ExchangeRate_Norm'] = (df_env['GBP'] - df_env['GBP'].mean()) / df_env['GBP'].std()
    df_env['Volatility'] = df_env['GBP'].rolling(window=30).std().fillna(0)

    # Encontrar índice de crisis
    try:
        crisis_date = pd.to_datetime('1992-09-16')
        crisis_idx = df_env[df_env['Date'] == crisis_date].index[0]
    except IndexError:
        # Si cae en fin de semana y no hay dato, buscamos el viernes anterior
        crisis_idx = df_env[df_env['Date'] < pd.to_datetime('1992-09-17')].index[-1]

    return df_env, crisis_idx

# --- PRUEBA ---
df_env, CRISIS_INDEX = prepare_real_data(df_cleaned, df_reserves)

print("--- Muestra de datos fusionados ---")
print(df_env[['Date', 'GBP', 'Total Reserves']].iloc[100:105])
```

```python
import random
from deap import base, creator, tools, algorithms

# Convertimos la fecha objetivo a Timestamp para comparaciones directas
TARGET_DATE = pd.to_datetime('1992-09-16')

# ==========================================
# 2. CONFIGURACIÓN DE DEAP
# ==========================================

if "FitnessMax" in creator.__dict__:
    del creator.FitnessMax
if "Individual" in creator.__dict__:
    del creator.Individual

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_float", random.uniform, -1.0, 1.0)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=3)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# ==========================================
# 3.FUNCIÓN DE EVALUACIÓN
# ==========================================

def eval_safety_robust(individual):
    w_reserves, w_rate, w_vol = individual
    threshold = 0.5
    
    # Calculamos la señal vectorizada
    signals = (w_reserves * df_env['Reserves_Norm']) + \
              (w_rate * df_env['ExchangeRate_Norm']) + \
              (w_vol * df_env['Volatility'])
    
    # Filtramos los días donde la señal supera el umbral
    trigger_indices = signals[signals > threshold].index
    
    if len(trigger_indices) == 0:
        # PENALIZACIÓN: Nunca salió del mercado, se comió la crisis.
        return -100.0,
    
    # Tomamos el PRIMER día que saltó la alarma
    first_exit_idx = trigger_indices[0]
    exit_date = df_env.loc[first_exit_idx, 'Date']
    
    # --- CÁLCULO DE FITNESS BASADO EN FECHAS (NO ÍNDICES) ---
    
    # Caso 1: Salió DESPUÉS de la crisis (Tarde)
    if exit_date > TARGET_DATE:
        # Penalización severa
        return -50.0,
        
    # Caso 2: Salió ANTES de la crisis
    else:
        # Calculamos cuántos días de antelación
        days_before = (TARGET_DATE - exit_date).days
        
        if days_before == 0:
            return 150.0, # ¡Jackpot! El mismo día (muy difícil)
            
        elif 1 <= days_before <= 14:
            # Zona Óptima (Danger Theory): 2 semanas antes
            # Puntuación: 100 puntos, menos 2 puntos por cada día de "miedo anticipado"
            # Si sale 1 día antes: 98 pts. Si sale 14 días antes: 72 pts.
            return 100.0 - (days_before * 2),
            
        elif 15 <= days_before <= 60:
            # Zona de Precaución: 1-2 meses antes
            # Puntuación positiva pero menor (50-60 pts)
            return 60.0 - (days_before * 0.5),
            
        else:
            # Falsa Alarma (Salió en 1991 o muy pronto en 1992)
            # Penalización por coste de oportunidad (perdió intereses)
            # Cuanto más pronto salga, peor la nota.
            return -days_before * 0.1, 

toolbox.register("evaluate", eval_safety_robust)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=0.2, indpb=0.2) # Mutación un poco más agresiva
toolbox.register("select", tools.selTournament, tournsize=3)

# ==========================================
# 4. EJECUCIÓN
# ==========================================

def main_robust():
    random.seed(42)
    # Aumentamos población y generaciones para que tenga tiempo de aprender
    pop = toolbox.population(n=100) 
    hof = tools.HallOfFame(1)
    
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("avg", np.mean)
    stats.register("max", np.max)
    
    print("Iniciando evolución robusta...")
    pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=40, 
                                   stats=stats, halloffame=hof, verbose=True)
    return pop, log, hof

# --- CORRER Y VISUALIZAR ---
pop, log, hof = main_robust()
best_ind = hof[0]

print(f"Pesos: Reservas={best_ind[0]:.2f}, Precio={best_ind[1]:.2f}, Volatilidad={best_ind[2]:.2f}")

# Visualizar
best_signal = (best_ind[0] * df_env['Reserves_Norm']) + \
              (best_ind[1] * df_env['ExchangeRate_Norm']) + \
              (best_ind[2] * df_env['Volatility'])

# Buscar fecha de salida
trigger = best_signal[best_signal > 0.5]

plt.figure(figsize=(14, 7))
plt.plot(df_env['Date'], df_env['GBP'], label='GBP Real', alpha=0.6, color='gray')

# Línea Roja: Miércoles Negro
plt.axvline(TARGET_DATE, color='red', linestyle='--', linewidth=2, label='Miércoles Negro (Real)')

if not trigger.empty:
    exit_idx = trigger.index[0]
    exit_date = df_env.loc[exit_idx, 'Date']
    exit_price = df_env.loc[exit_idx, 'GBP']
    
    plt.axvline(exit_date, color='green', linewidth=2, label='Señal Danger Theory')
    plt.plot(exit_date, exit_price, 'go', markersize=12)
    
    print(f"El agente salió el: {exit_date.date()}")
    print(f"Días de antelación: {(TARGET_DATE - exit_date).days}")
    

plt.title('Predicción Danger Theory (Ajustada)')
plt.legend()
plt.grid(True)
plt.xlim(pd.Timestamp('1992-01-01'), pd.Timestamp('1993-01-01')) # Zoom en el año clave
plt.show()
```
